# Salary Distribution

```python
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading Data
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df['job_skills']=df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_ind=df[(df['job_country']=='India')].dropna(subset=['salary_year_avg'])
df_ind
```

```python
job_titles=df_ind['job_title_short'].value_counts().index[:6].tolist()
job_titles
```

```python
df_ind_top6=df_ind[df_ind['job_title_short'].isin(job_titles)]
df_ind_top6
```

```python
job_order=df_ind_top6.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False).index
```

```python
sns.boxplot(x='salary_year_avg', y='job_title_short', data=df_ind_top6, order=job_order)
plt.xlabel('Average Salary')
plt.ylabel('') 
plt.title('Salary Distribution of Data Jobs in the India')
ax= plt.gca()
ticks_x=plt.FuncFormatter(lambda x, pos: f'₹{int(x*87.5/100000)}L')
plt.gca().xaxis.set_major_formatter(ticks_x)
plt.xlim(0, 300000)
plt.show()
```

### Median Salary Vs Skills for Data Scientists 

```python

df_ind2=df[(df['job_title_short']=='Data Scientist')&(df['job_country']=='India')].copy()

df_ind2=df_ind2.dropna(subset=['salary_year_avg'])
df_ind2
```

```python
df_ind2=df_ind2.explode('job_skills')
df_ind2[['job_skills','salary_year_avg']].head(5)
```

```python
df_ind_top_pay=df_ind2.groupby('job_skills')['salary_year_avg'].agg(['count','median']).sort_values(by='median',ascending=False).head(10)
df_ind_top_pay
```

```python
df_ind_skills=df_ind2.groupby('job_skills')['salary_year_avg'].agg(['count','median']).sort_values(by='count',ascending=False).head(10)
df_ind_skills=df_ind_skills.sort_values(by='median',ascending=False)
df_ind_skills
```

```python
sns.set_theme(style='ticks')
fig, ax=plt.subplots(2,1)
sns.barplot(data=df_ind_top_pay, y=df_ind_top_pay.index, x='median', ax=ax[0], hue='median',palette='dark:b_r')
ax[0].legend().set_visible(False)
ax[0].set_ylabel('')
ax[0].set_xlabel('')
ax[0].set_title('Top 10 Payed Skills in India')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'₹{int(x*87.5/100000)}L'))

sns.barplot(data=df_ind_skills, y=df_ind_skills.index, x='median', ax=ax[1], hue='median',palette='light:b')
ax[1].legend().set_visible(False)
ax[1].set_xlim(ax[0].get_xlim())
ax[1].set_ylabel('')
ax[1].set_xlabel('Median Salary')
ax[1].set_title('Top 10 Demanded Skills in India')
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'₹{int(x*87.5/100000)}L'))

plt.tight_layout()
plt.show()
```
